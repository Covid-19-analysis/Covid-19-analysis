```python
us_vaccination = "./resources/us_state_vaccinations.csv"

us_vaccination_df = pd.read_csv(us_vaccination)
us_vaccination_df

us_vaccination_df = us_vaccination_df.dropna(subset = ['total_distributed', 'people_fully_vaccinated', 'daily_vaccinations_raw'])
us_vaccination_df
```

```python
vaccination_m_df = "./resources/vaccinations-by-manufacturer.csv"

vaccination_m_df = pd.read_csv(vaccination_m_df)

vaccination_us_df = vaccination_m_df.loc[vaccination_m_df["location"] == "United States"]
vaccination_us_df
```

```python
vaccination_usa_df = vaccination_us_df.dropna(subset = ['vaccine', 'total_vaccinations'])
vaccination_usa_df
```

```python
date = (vaccination_usa_df['date']> '2021-01-01') & (vaccination_usa_df['date']< '2021-02-01')
jan_vaccination_df = vaccination_usa_df.loc[date].groupby(['vaccine'])

jan_vaccination_df.count()
```

```python
date = (vaccination_usa_df['date']> '2021-02-01') & (vaccination_usa_df['date']< '2021-03-01')
feb_vaccination_df = vaccination_usa_df.loc[date].groupby(['vaccine'])

feb_vaccination_df.count()
```
